13. Why is memory management important in Python?
 - Automatic garbage collection: Python tracks objects and automatically clears memory when they're no longer in use using reference counting and the gc (garbage collector) module.
  - Efficient memory allocation: Python’s internal memory manager handles object allocation through private heaps, which keep memory organized and avoid fragmentation.
  - Reference counting: Each object keeps a count of references to it—when that count hits zero, the memory is eligible for release.
  - Support for dynamic types: Python’s flexibility with types (e.g., integers, lists, dictionaries) means memory needs can vary widely, and good management helps avoid waste.

15. How does Python's garbage collection system work?
- Python’s garbage collection (GC) system automatically cleans up memory by deleting objects that are no longer needed, freeing up space for new objects.

 1. Reference Counting:  
  - Every object in Python keeps track of how many references point to it.

  - When the reference count drops to zero, the object is immediately deleted.
  2. Cyclic Garbage Collector:
  - Sometimes two or more objects reference each other (a circular reference) but are no longer used.

  - Python’s GC module detects such reference cycles and cleans them up.
  3. Generational Garbage Collection:
  - Python groups objects into three generations (0, 1, and 2).

  - New objects → Gen 0

  - Survive longer → Move to higher generations

  - Frequent cleanup is done on young objects (Gen 0) for performance.

19. What is the importance of closing a file in Python?
 - When you open a file in Python (using open()), the system allocates resources to handle that file. Closing the file is essential to release those resources and ensure data integrity.
 - Importance:
     1. Frees System Resources:
       - Open files consume system-level resources (file descriptors).
     2. Ensures Data Is Written (in Write Modes):
       - When writing ("w", "a"), data is first stored in a buffer.
       - Closing the file flushes this buffer to disk — making sure all data is actually saved.
      3. Prevents File Corruption:
        - Especially important in write mode — failing to close may result in partial or corrupt files.
      4. Allows Other Programs to Access the File:
        - Some systems lock a file when it’s open.

        - Closing it allows other programs or users to safely access or modify it.

23.  What are the challenges associated with memory management in Python?
  1. Garbage Collection Complexity
      Python uses automatic memory management via a garbage collector.

      The cyclic garbage collector detects and handles reference cycles—but it's not instant and can cause performance hiccups.

      Developers might need to manually tune or trigger garbage collection using the gc module for optimization

  2. Reference Counting
      Every object has a reference count; when it drops to zero, memory is freed.

      But circular references (e.g., objects referencing each other) can prevent counts from reaching zero, leading to memory leaks if not handled.

  3. Memory Leaks
      Unintended retention of objects—especially in large applications or long-running scripts—can consume more memory than expected.

      Leaks often stem from global references, closures, or lingering exceptions holding onto frames.

  4. Lack of Fine-Grained Control
      Python abstracts away many low-level memory operations.

      Unlike languages like C, developers can't directly allocate or free memory, which can frustrate those needing precision for performance.

  6. Third-Party Library Usage
      Libraries with native extensions (e.g. NumPy, TensorFlow) interact with memory differently, and careless usage may bypass Python's safety mechanisms.

```python
#4.Write a Python script that reads from one file and writes its content to another file.
# File: copy_file.py

# Source and destination file paths
source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open the source file for reading
    with open(source_file, "r") as src:
        content = src.read()

    # Open the destination file for writing
    with open(destination_file, "w") as dest:
        dest.write(content)

    print(f"Contents copied from {source_file} to {destination_file} successfully.")

except FileNotFoundError:
    print(f"The file {source_file} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
#6.Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

# Configure the logging system
logging.basicConfig(
    filename="error.log",  # Log file name
    level=logging.ERROR,   # Only log ERROR level and above
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# Function to perform division
def divide(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError as e:
        logging.error("Division by zero error: %s", e)
        print("Error: Cannot divide by zero.")

# Example usage
num1 = 10
num2 = 0

divide(num1, num2)
```

```python
#14.Write a program that uses the logging module to log both informational and error messages?
import logging

# Configure logging
logging.basicConfig(
    filename="app.log",            # Log file name
    level=logging.INFO,            # Minimum logging level
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# Example function
def divide(a, b):
    logging.info(f"Attempting to divide {a} by {b}")
    try:
        result = a / b
        logging.info(f"Division successful: {a} / {b} = {result}")
        return result
    except ZeroDivisionError as e:
        logging.error(f"Error occurred: {e}")
        return None

# Example usage
divide(10, 2)   # Logs INFO
divide(10, 0)   # Logs ERROR
```

```python
#15. Write a Python program that prints the content of a file and handles the case when the file is empty?
def read_file(file_path):
    try:
        with open(file_path, "r") as file:
            content = file.read()

            if content.strip() == "":
                print(f"The file '{file_path}' is empty.")
            else:
                print(f"Contents of '{file_path}':\n")
                print(content)

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
file_name = "example.txt"
read_file(file_name)
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)  # You can choose INFO, WARNING, etc.

# Set up RotatingFileHandler
handler = RotatingFileHandler(
    "app.log",        # Log file name
    maxBytes=1_000_000,  # Rotate after 1MB (1,000,000 bytes)
    backupCount=3        # Keep 3 backup log files (e.g., app.log.1, app.log.2, etc.)
)

# Log format
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add handler to logger
logger.addHandler(handler)

# Example logging
for i in range(10000):
    logger.info(f"This is log entry number {i}")
```

```python
#19.Write a program that handles both IndexError and KeyError using a try-except block?
def handle_errors():
    my_list = [1, 2, 3]
    my_dict = {"name": "Alice", "age": 25}

    try:
        # IndexError example: access out-of-range index
        print("List item:", my_list[5])

        # KeyError example: access missing key
        print("City:", my_dict["city"])

    except IndexError as e:
        print(f"IndexError caught: {e}")

    except KeyError as e:
        print(f"KeyError caught: The key '{e}' does not exist in the dictionary.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Run the function
handle_errors()
```
